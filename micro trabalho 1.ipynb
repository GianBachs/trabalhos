```python
import pandas_datareader as web
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
a = float(input('Digite o numero da variável a: '))
b = float(input('Digite o numero da variável b [0.1-0.9]: '))
c = float(input('Digite o numero da variável c: '))
d = float(input('Digite o numero da variável d [0.1-0.9]: '))

#Variaveis Endógenas


preco = (c - a) / (b + d)

if preco <0:
      print ('Preço negativo, digite outro valor de parâmetros')

elif (b + d) == 0: 
      print ('digite outro valor de parâmetros') 

else: 
      print(f'O preço de equilíbrio é igual a: {preco}') #fstring

qo = a + b * preco

print(f'A quantidade de equilibrio é igual a: {qo}')

df= [qo, preco]
```

```python

plt.figure(figsize=(18,6))
plt.title('Função oferta e demanda')
plt.plot(df)

plt.xlabel('Preço', fontsize=18)
plt.ylabel('Quantidade de equilibrio', fontsize=18)
plt.show
```
